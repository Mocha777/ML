```python
# 初始化參數
X = YearsExperience
Y = Salary
alpha = 0.001  # 學習率
epochs = 20000  # 迭代次數
n = len(YearsExperience)  # 資料數量
# 儲存損失值以便畫圖
loss_history = []

# 梯度下降
for epoch in range(epochs):
    y_pred = m * X + b  # 預測值
    error = Y - y_pred  # 殘差

    # 計算梯度
    dm = (-2 / n) * np.sum(X * error)
    db = (-2 / n) * np.sum(error)

    # 更新參數
    m = m - alpha * dm
    b = b - alpha * db

    # 計算 MSE
    loss = np.mean(error ** 2)
    loss_history.append(loss)

    # 每 100 次輸出一次進度
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.2f}, m = {m:.2f}, b = {b:.2f}")

# 繪製損失函數變化圖
plt.figure(figsize=(8, 5))
plt.plot(range(epochs), loss_history, label="Loss", color="blue")
plt.xlabel("Epoch")
plt.ylabel("Loss (MSE)")
plt.title("Loss Reduction Over Time (Gradient Descent)")
plt.grid(True)
plt.legend()
plt.show()

# 繪製最佳擬合線
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color="blue", label="Actual Data", alpha=0.6, edgecolors="black")
plt.plot(X, m * X + b, color="red", label=f"Best Fit Line: y = {m:.2f}x + {b:.2f}")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression with Gradient Descent.")
plt.legend()
plt.grid(True)
plt.show()

# 最終結果
print(f"Final model: y = {m:.2f}x + {b:.2f}")
```
